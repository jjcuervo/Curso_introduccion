I. Describir el problema de la organización:


El Banco Falabella es una de las entidades financieras más importantes en términos de colocación de tarjetas de crédito en Colombia. Esta entidad maneja dos tipos de clientes. El primer grupo es el grupo de clientes que frecuentan el ecosistema y son evaluados de manera batch (proceso masivo) y se genera una pre aprobación al evaluar estos clientes por un modelo de machine learning y diferentes criterios de riesgo de crédito. El segundo segmento, representa a los clientes que no han tenido ningún tipo de contacto con el banco o sus aliados (ecosistema). Actualmente, el banco cuenta con un modelo de baja predicción para evaluar este segmento de clientes y su tasa de aprobación es del 0,5% (se aprueban aproximadamente 500 tarjetas de 100.000 solicitudes evaluadas). Desde las áreas de análisis de riesgo de crédito, uno de los retos más relevantes es la correcta clasificación de clientes potenciales entre quienes probablemente cumplirán sus obligaciones crediticias (buenos) y quienes tienen un alto riesgo de impago (malos).

La baja aprobación de este segmento está limitando al banco a no poder potenciar su crecimiento, rechazando posibles clientes buenos para la entidad, dejando de maximizar el potencial y generando posibles pérdidas reputacionales.Un error en esta clasificación puede representar pérdidas financieras por mora o, en sentido contrario, pérdida de oportunidades comerciales por rechazos innecesarios. Además, áreas como originación de crédito, cobranza y comercial se ven directamente impactadas por la calidad de esta clasificación.

II. Describir cómo la ciencia de datos apoyada con IA puede ayudar a resolver el reto:

La ciencia de datos apoyada por IA puede solucionar este problema al desarrollar un modelo que pueda predecir de una mejor manera el comportamiento de los clientes del segmento de espontáneos para maximizar la aprobación de los mismos.

El modelo a utilizar es un modelo de regresión logística que nos permita separar la población entre buenos y malos clientes de acuerdo a ciertas características en común.

El uso de Ciencia de Datos e Inteligencia Artificial es especialmente pertinente en este contexto por varias razones:

Volumen y variedad de datos: A diario se evalúan miles de solicitudes, cada una con múltiples variables provenientes de burós de crédito, comportamiento interno, sector económico, y variables socioeconómicas. Procesar esto de forma manual o con reglas rígidas es ineficiente.

Técnicas de clasificación supervisada permiten desarrollar modelos que aprenden patrones complejos que correlacionan variables históricas con el comportamiento de pago. Esto aumenta la precisión y confiabilidad en la toma de decisiones.

Modelos recomendados:

Logistic Regression: Modelo base, interpretable, ideal para evaluar la probabilidad de default y entender la influencia individual de cada variable.

Random Forest Classifier: Modelo robusto no lineal que maneja muy bien interacciones entre variables, es resistente al sobreajuste y provee importancia de variables.

y uno más, el modelo XGBoost es una opción más sofisticada para producción, con alta precisión y capacidad para manejar desbalance de clases.

Beneficios específicos:

Mejora en los niveles de aprobación sin aumentar riesgo.

Reducción de mora temprana y provisiones.

Agiliza la originación y mejora la experiencia del cliente.

Metodología CRISP-DM:

En este caso, la aplicación de CRISP-DM se adapta perfectamente, comenzando con la comprensión del negocio (riesgo de impago), comprensión de los datos (historial de clientes), preparación, modelado, evaluación y finalmente, despliegue en la plataforma de originación.

Para este ejercicio voy a escoger la variable cuotas. La variable mora solicitud tiene mejor correlación pero esta variable puede tener problemas de causalidad con la variable objetivo dado que esa mora al momento de la solicitud puede ser la misma mora de la variable resultado a lo largo del tiempo. Para ajustar esta parte debemos separar la población que no tenía mora al momento de la solicitud de la población que si tenía mora al momento de la solicitud.

La variable cuotas puede influir de manera importante en la mora de los clientes. Al tener que pagar una cuota muy alta, el cliente puede verse en situaciones de poca liquidez, generando entrada en mora en los pagos de sus obligaciones.

```python
from scipy.stats import chi2_contingency

for col in df.columns.drop(target):
    table = pd.crosstab(df[col], df[target])
    chi2, p, dof, expected = chi2_contingency(table)
    print(f"{col}: chi2={chi2:.2f}, p-valor={p:.4f}")
```

```python
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r,k = confusion_matrix.shape
    return np.sqrt(chi2 / (n*(min(k,r)-1)))

for col in df.columns.drop(target):
    table = pd.crosstab(df[col], df[target])
    cv = cramers_v(table)
    print(f"{col}: Cramér’s V = {cv:.3f}")
```

```python
## apilamos al 100% gráfico de barras entre experiencia y moras

import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# Crosstab (ejemplo Exp abierta vs Mora sector)
tabla = pd.crosstab(index=df["Exp abierta"], columns=df["mora sector"])

# Normalizar cada fila para que sume 1 (100%)
tabla_pct = tabla.div(tabla.sum(axis=1), axis=0)

# Gráfico apilado al 100%
ax = tabla_pct.plot(kind="bar", stacked=True, figsize=(8,4), colormap="Set2")
ax.set_title("Mora por Exp abierta (apilado 100%)")
ax.set_xlabel("Exp abierta")
ax.set_ylabel("Porcentaje")
ax.yaxis.set_major_formatter(PercentFormatter(1.0)) # mostrar eje Y en %

plt.xticks(rotation=0)
plt.legend(title="Mora sector", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()

# etiquetas de porcentaje (opcional)
for container in ax.containers:
    labels = [f"{v*100:.0f}%" if v>0 else "" for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type="center")

plt.show()
```
